```python
page = requests.get("https://www.tourlane.fr/asie/malaisie/sites-touristiques/#parc-national-de-taman-negara", headers=headers)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
h1 = soup.findAll('h1')
h2 = soup.findAll('h2')
h3 = soup.findAll('h3')
h4 = soup.findAll('h4')
h5 = soup.findAll('h5')
h6 = soup.findAll('h6')
titles = [h1,h2,h3,h4,h5,h6]
```

```python
list_of_titles = []
for title in titles:
    for n in range(len(title)):
        list_of_titles.append(title[n].text)
```

```python
input_text = ' '.join(list_of_titles)
cleaned_text = clean_text(str(input_text))
new = remove_duplicates(cleaned_text)
new = preprocess_text(new)
```
